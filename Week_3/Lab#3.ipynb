# Step 2: Handle Missing Values

```python
# Step 2: Handle Missing Values
# Decide how to handle missing values based on context:
# Example: Drop rows where 'DATE' is missing (or replace with a default value)
df.dropna(subset=['DATE'], inplace=True)  # Drop rows with missing DATE

# For other columns, you can fill missing values with appropriate defaults:
df['BRIDGE IDENTIFICATION NUMBER'] = df['BRIDGE IDENTIFICATION NUMBER'].fillna(0)  # Replace NaN with 0
df['MAIN STREET'] = df['MAIN STREET'].fillna('Unknown')  # Replace NaN with 'Unknown'
```

# Step 3: Convert Data Types (as per your earlier request)


```python
df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')  # Convert to datetime
df['BOROUGH'] = df['BOROUGH'].astype('category')          # Convert to category
df['DIRECTION'] = df['DIRECTION'].astype('category')      # Convert to category
df['MAIN STREET'] = df['MAIN STREET'].astype('string')    # Convert to string
df['CROSS STREET'] = df['CROSS STREET'].astype('string')  # Convert to string
df['TYPE'] = df['TYPE'].astype('category')                # Convert to category
df['TRUCK ROUTE'] = df['TRUCK ROUTE'].astype('category')  # Convert to category
df['STRUCK OVERPASS'] = df['STRUCK OVERPASS'].astype('category')  # Convert to category
df['STATE OF OPERATORS LICENSE'] = df['STATE OF OPERATORS LICENSE'].astype('string')  # Convert to string
df['BRIDGE IDENTIFICATION NUMBER'] = df['BRIDGE IDENTIFICATION NUMBER'].astype('Int64')  # Nullable integer
df['LATITUDE'] = pd.to_numeric(df['LATITUDE'], errors='coerce')  # Ensure float type
df['LONGITUDE'] = pd.to_numeric(df['LONGITUDE'], errors='coerce')  # Ensure float type
```

```python
df.dtypes
```

# Step 4: Extract Day, Month, and Year from DATE

```python
df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce') 
```

```python
df['Day'] = df['DATE'].dt.day       # Extract day
df['Month'] = df['DATE'].dt.month   # Extract month
df['Year'] = df['DATE'].dt.year     # Extract year
```

```python
df.head()
```

# Step 5: Handle Duplicates (if any)

```python
print("\nChecking for Duplicates...")
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
if duplicate_count > 0:
    df.drop_duplicates(inplace=True)
    print("Duplicates removed.")
```

# Step 6: Validate Latitude and Longitude Ranges (if applicable)



# Lab #4 

```python
# Latitude should be between -90 and 90, Longitude between -180 and 180.
invalid_latitude = ((df['LATITUDE'] < -90) | (df['LATITUDE'] > 90)).sum()
invalid_longitude = ((df['LONGITUDE'] < -180) | (df['LONGITUDE'] > 180)).sum()
print(f"\nInvalid Latitude Count: {invalid_latitude}")
print(f"Invalid Longitude Count: {invalid_longitude}")
```

# Optionally drop invalid rows:

```python

df = df[(df['LATITUDE'] >= -90) & (df['LATITUDE'] <= 90)]
df = df[(df['LONGITUDE'] >= -180) & (df['LONGITUDE'] <= 180)]
```

 # Data Visualization

```python


# Bar plot
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
df['BOROUGH'].value_counts().plot(kind='bar')
plt.title('Distribution of Incidents by Borough')
plt.xlabel('Borough')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
#Truck Route vs Non-Truck Route (Pie Chart)

plt.figure(figsize=(8, 8))
df['TRUCK_ROUTE'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Truck Route vs Non-Truck Route')
plt.ylabel('')
plt.show()
```

```python
#Top 10 States of Operators License (Bar Chart)
plt.figure(figsize=(10, 6))
df['STATE_OF_OPERATORS_LICENSE'].value_counts().nlargest(10).plot(kind='bar')
plt.title('Top 10 States of Operators License')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Incidents Over Time (Line Plot)
df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE').resample('M')['BRIDGE_IDENTIFICATION_NUMBER'].count().plot(figsize=(12, 6))
plt.title('Number of Incidents Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Incidents')
plt.tight_layout()
plt.show()
```

```python
# Heatmap of Incidents by Month and Day
pivot = df.pivot_table(index='Month', columns='Day', aggfunc='size', fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(pivot, cmap='YlOrRd')
plt.title('Heatmap of Incidents by Month and Day')
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of each unique value in TRUCK_ROUTE
truck_route_counts = df['TRUCK_ROUTE'].value_counts()

# Bar plot for TRUCK_ROUTE distribution
plt.figure(figsize=(8, 5))
truck_route_counts.plot(kind='bar', color=['skyblue', 'orange'])

# Add title and labels
plt.title('Distribution of Truck Route (YES vs. NO)', fontsize=14)
plt.xlabel('Truck Route', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Annotate bar values
for index, value in enumerate(truck_route_counts):
    plt.text(index, value + 1, str(value), ha='center', fontsize=10)

# Show the plot
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
