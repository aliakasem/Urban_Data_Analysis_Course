```python


# Introduction to Linear Regression 

# --- Lesson Outline ---
# 1. Introduction to Regression
# 2. Assumptions of Linear Regression
# 3. Implementing Regression in Python
# 4. Visualizing Regression and Residuals

# --- 1. Introduction to Regression ---
# Regression analysis estimates the relationship between a dependent variable and one or more independent variables.
# Simple linear regression has one independent variable (X) and one dependent variable (Y).
# Equation: Y = b0 + b1*X + e

# --- 2. Assumptions of Linear Regression ---
# a. Linearity: Relationship between X and Y is linear.
# b. Independence: Observations are independent.
# c. Homoscedasticity: Constant variance of residuals.
# d. Normality: Residuals are normally distributed.
# e. No multicollinearity (for multiple regression).

# --- 3. Implementing Regression in Python ---

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Create synthetic dataset
np.random.seed(42)
X = 2.5 * np.random.randn(100) + 1.5   # Feature
residuals = 0.5 * np.random.randn(100)       # Random noise
Y = 2 + 0.3 * X + residuals          # Target variable

# Reshape and split data
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train model
model = LinearRegression()
model.fit(X_train, Y_train)

# Predict
Y_pred = model.predict(X_test)

# --- 4. Visualizing Regression and Residuals ---

# Plot regression line
plt.figure(figsize=(10, 6))
plt.scatter(X_test, Y_test, color='blue', label='Actual')
plt.plot(X_test, Y_pred, color='red', linewidth=2, label='Predicted Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Line')
plt.legend()
plt.show()

# Plot residuals
residuals = Y_test - Y_pred
plt.figure(figsize=(10, 6))
plt.scatter(Y_pred, residuals, color='purple')
plt.hlines(y=0, xmin=min(Y_pred), xmax=max(Y_pred), color='black', linewidth=2)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()

# --- Evaluation ---
print(f"Mean Squared Error: {mean_squared_error(Y_test, Y_pred):.3f}")
print(f"R-squared: {r2_score(Y_test, Y_pred):.3f}")
```
